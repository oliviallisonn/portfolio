```python
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup

page = requests.get("https://en.wikipedia.org/wiki/Sour_(album)")
soup = BeautifulSoup(page.text,"html")
sour_table = soup.find_all("table")[5]
sour_data = sour_table.find_all("td")
sour_tracks = [track.text.strip().strip("\"") for track in sour_data if track.text.strip() != "" and track.text.strip()[0] == "\""]
sour_spotify_streams = [527086697,1277052620,1905590323,377104314,1307274925,1930339344,378709165,891022443,694027856,819312506,274304445]
```

```python
plt.barh(sour_tracks,sour_spotify_streams,color="#d3bee8")
plt.title("Olivia Rodrigo SOUR Streams on Spotify (By Millions)(Sept 2023)")
```
