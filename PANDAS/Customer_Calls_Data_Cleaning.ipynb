```python
import pandas as pd

df = pd.read_excel("/Users/olivia/Downloads/Customer Call List.xlsx")
df
```

```python
#CLEANING LAST NAME COLUMN
df["Last_Name"] = df["Last_Name"].str.strip("/._")
```

```python
#REMOVING DUPLICATE ROWS
df.drop_duplicates(inplace=True)
```

```python
#REMOVING ROWS WITH NULL ANSWERS IN PHONE NUMBER AND ADDRESS COLUMNS
df.fillna("NaN", inplace=True)
null_values = ["NaN", "N/a"]
df.drop(index=df[df["Phone_Number"].isin(null_values)].index,inplace=True)
df.drop(index=df[df["Address"].isin(null_values)].index, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
#CLEANING PHONE NUMBER COLUMN TO FOLLO 123-456-7890 FORMAT
df["Phone_Number"] = df["Phone_Number"].astype("string")
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x.replace("-","").replace("|","").replace("/",""))
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3]+ "-" + x[3:6] + "-" + x[6:9])
df["Phone_Number"]
```

```python
df
```

```python
#NOTICED THAT PAYING CUSTOMER AND CUSTOMER ID COLUMNS WEREN'T UNIFORMED WITH OTHER COLUMN NAMES
df.rename(columns={"Paying Customer":"Paying_Customer","CustomerID" : "Customer_ID"},inplace=True)
df
```

```python
#STANDARDIZING RESPONSES IN PAYING CUSTOMER AND DO NOT CONTACT COLUMNS
df["Paying_Customer"].replace({"Yes": "Y","No":"N"},inplace=True)
df["Do_Not_Contact"].replace({"Yes": "Y","No":"N","NaN": "N"},inplace=True)
```

```python
df
```

```python
#DROPPING NOT USEFUL COLUMN
df.drop(columns="Not_Useful_Column",inplace=True)
```

```python
#ONLY SELECTING CUSTOMERS THAT ARE OKAY WITH BEING CONTACTED
df = df.loc[df["Do_Not_Contact"] == "N"]
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
#DROPPING DO NOT CONTACT COLUMN BECAUSE IT'S NOT USEFUL ANYMORE
df.drop(columns="Do_Not_Contact",inplace=True)
```
