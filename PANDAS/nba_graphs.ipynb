```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
df = pd.read_csv("nba_games.csv")
df.head()  
```

```python
fran_count = Counter(df["fran_id"])
opp_count = Counter(df["opp_fran"])
opp_final = {}
```

```python
fran_count
```

```python
for k,v in opp_count.items():
    if k in fran_count:
        opp_final[k] = v
        
opp_final
```

```python
teams = [key for key in fran_count.keys()]
teams
```

```python
fran_num = [v for v in fran_count.values()]
fran_num
```

```python
opp_num = [v for v in opp_final.values()]
opp_num
```

```python
import numpy as np
team_np = np.arange(len(teams))
width = 0.25
plt.bar(teams, fran_num, color = "r", width = 0.25,label= "Main Franchise")
plt.bar(team_np + width, opp_num, color="b", width=0.25, label = "Opponent Franchise")
plt.xlabel("Teams")
plt.ylabel("Frequency")
plt.tight_layout()
plt.legend()
```

```python
knicks_years = df.groupby("fran_id").get_group("Knicks")["year_id"]
knicks_years_counter = Counter(knicks_years)
years = [k for k in knicks_years_counter.keys()]
knicks_count = [v for v in knicks_years_counter.values()]
knicks_years_counter
```

```python
celtics_years = df.groupby("fran_id").get_group("Celtics")["year_id"]
celtics_years_counter = Counter(celtics_years)
celtics_count = [v for v in celtics_years_counter.values()]
celtics_years_counter
```

```python
plt.plot(years,knicks_count,color="r",label="Knicks")
plt.xlabel("Years")
plt.ylabel("Frequency")
plt.grid()
plt.plot(years,celtics_count,color="b",label="Celtics")
plt.legend()
plt.tight_layout()
plt.title("Knicks and Celtics Games")
```

```python
thunder_years = df.groupby("fran_id").get_group("Thunder")["year_id"]
thunder_years_counter = Counter(thunder_years)
thunder_count = [v for v in thunder_years_counter.values()]
years = [k for k in thunder_years_counter.keys()]
len(thunder_years_counter)
```

```python
spurs_years = df.groupby("fran_id").get_group("Spurs")["year_id"]
spurs_years_counter = Counter(spurs_years)
spurs_count = [v for v in spurs_years_counter.values()]
len(spurs_years_counter)
```

```python
nets_years = df.groupby("fran_id").get_group("Nets")["year_id"]
nets_years_counter = Counter(nets_years)
nets_count = [v for v in nets_years_counter.values()]
len(nets_years_counter)
```

```python
plt.plot(years,thunder_count,color="k",label="Thunder")
plt.plot(years,spurs_count,color="r",label="Spurs")
plt.plot(years,nets_count,color="b",label="Nets")
plt.title("Thunder, Spurs, Nets Games")
plt.xlabel("Years")
plt.ylabel("Frequency")
plt.grid()
plt.legend()
```
