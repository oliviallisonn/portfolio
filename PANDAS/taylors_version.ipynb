```python
import requests 
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt

#FEARLESS TV
page = requests.get("https://en.wikipedia.org/wiki/Fearless_(Taylor%27s_Version)")
soup = BeautifulSoup(page.text,"html")
tables = soup.find_all("table")
tracklist_table = tables[3]
tracklist_data = tracklist_table.find_all("td")
tracklist = [track.text.strip().replace("\"","") for track in tracklist_data if track.text.strip()[0] == "\""]
streams = [165534082,75114139,491267082,67760682,70345226,405207961,51023529,43810656,53945181,307619468,82329622,39438297,34261889,41347370,41815627,39584092,31047205,27973972,54123544,40327433,93394529,283574271,45923615,61564595,53784607,42509618]
final_data = sorted([(streams,track) for track,streams in zip(tracklist,streams)])
fearless_most_streamed = final_data[-1]
fearless_tracks = [v for k,v in final_data]
fearless_streams = [k for k,v in final_data]
```

```python
#FEALRESS BAR GRAPH
plt.barh(fearless_tracks,fearless_streams,color="#ddc477")
plt.title("Fearless (TV) Spotify Streams Sept 2023 (By Hundred Millions)")
```

```python
#SPEAK NOW
page = requests.get("https://en.wikipedia.org/wiki/Speak_Now_(Taylor%27s_Version)")
soup = BeautifulSoup(page.text, "html")
tables = soup.find_all("table")
tracklist_table = tables[3]
tracklist_data = tracklist_table.find_all("td")
tracklist = [track.text.strip().replace("\"", "") for track in tracklist_data if track.text.strip()[0] == "\""]
streams = [74584430,56023464,82099216,51831364,50737863,45348900,45294974,29377821,94397944,58956049,25691645,36677952,35770811,47862579,26916752,21953578,43591612,42840269,120996401,43423124,43754751,40457669]
final_data = sorted([(v,k) for k,v in zip(tracklist,streams)])
speaknow_most_streamed = final_data[-1]
speaknow_tracks = [v for k,v in final_data]
speaknow_streams = [k for k,v in final_data]
```

```python
#SPEAK NOW BAR GRAPH
plt.barh(speaknow_tracks,speaknow_streams,color="#e2b7ce")
plt.title("Speak Now (TV) Spotify Streams Sept 2023 (By Hundred Millions)")
```

```python
#TOP SONGS PER ALBUM BAR GRAPH
most_streamed = [fearless_most_streamed, red_most_streamed, speaknow_most_streamed]
most_streamed_songs = [v for k,v in most_streamed]
most_streamed_streams = [k for k,v in most_streamed]
plt.bar(most_streamed_songs, most_streamed_streams, color="#b8acd1")
plt.title("Most Streamed Songs of Each Taylor's Version Album Sept 2023 (By Millions)")
```

```python
#ALBUMS SUMMARY
fearless_array = np.array(fearless_streams)
red_array = np.array(red_streams)
speaknow_array = np.array(speaknow_streams)
total_streams = [sum(fearless_array),sum(red_array),sum(speaknow_array)]
tv_albums = ["Fearless","Red","Speak Now"]
colors = ["#ddc477","#9c1b15","#e2b7ce"]

plt.pie(total_streams, labels=tv_albums,colors=colors,autopct="%1.1f%%",wedgeprops = {"edgecolor":"black"})
plt.title("Spotify Streaming Breakdown of Taylor's Version Albums")
```
