```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
df2 = pd.read_csv("/Users/olivia/Downloads/Untitled spreadsheet - ezGeocode #2.csv")
addresses = df2["Address"].tolist()
```

```python
categories = ['Total Lot Area (approx):','Address:','Title:', 'Property ID:','Lot Area (approx):','Suburb:','Local Government Area:','Federal Electorate District:','State Electorate District:','Statistical Area 2:','First Nation Countries:']
df = pd.DataFrame(columns=categories)
driver = webdriver.Chrome()
driver.get("https://meconemosaic.au/")
time.sleep(7)
nsw = driver.find_element(By.LINK_TEXT,"NSW")
nsw.click()
time.sleep(2)
close_button = driver.find_elements(By.CLASS_NAME,"MuiButtonBase-root")[29]
close_button.click()
time.sleep(3)
for address in addresses:
    search = driver.find_element(By.CLASS_NAME,"mapboxgl-ctrl-geocoder--input")
    search.clear()
    search.send_keys(address)
    time.sleep(3)
    search.click()
    search.send_keys(Keys.RETURN)
    time.sleep(5)
    categories_raw = driver.find_elements(By.CLASS_NAME,"info-click-layer-feature-attribute-name")
    categories_text = [category.text.strip() for category in categories_raw]
    data_raw = driver.find_elements(By.CLASS_NAME,"info-click-layer-feature-attribute-value")
    data_text = [data.text.strip() for data in data_raw]
    
    cat_data = {k:v for k,v in zip(categories_text,data_text)}
    row_dic = {}
    for category in categories:
        if category in cat_data:
            row_dic[category] = cat_data[category]
        else:
            row_dic[category] = 0
    
    temp_df = pd.DataFrame([row_dic.values()],columns=categories)
    df = pd.concat([df,temp_df])
    df.reset_index(inplace=True,drop=True)
df.insert(0,"Start",addresses)
df
```
